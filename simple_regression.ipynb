```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

# === Generišemo proizvoljne tačke radi ilustracije ===
np.random.seed(0)
x = np.linspace(0, 10, 30)
y = 2.5 * x + 5 + np.random.normal(0, 3, size=len(x))  # malo "šuma" za realističnost


# === Funkcija koja iscrtava liniju ===
def prikazi_regresiju(a=5.0, b=2.5):
    plt.figure(figsize=(7, 5))

    # tačke (posmatrani podaci)
    plt.scatter(x, y, color="blue", label="Podaci (tačke)")

    # regresiona linija
    x_lin = np.linspace(min(x), max(x), 200)
    y_lin = a + b * x_lin
    plt.plot(x_lin, y_lin, color="red", linewidth=2, label=f"y = {a:.2f} + {b:.2f}x")

    plt.title("Interaktivna regresiona linija")
    plt.xlabel("x (nezavisna varijabla)")
    plt.ylabel("y (zavisna varijabla)")
    plt.legend()
    plt.grid(True)
    plt.show()


# === Interaktivni slajderi ===
interact(
    prikazi_regresiju,
    a=FloatSlider(value=5, min=-10, max=10, step=0.5, description="a (odsečak)"),
    b=FloatSlider(value=2.5, min=-5, max=5, step=0.1, description="b (nagib)"),
)
```
